```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate sample data
np.random.seed(0)
xs = np.random.randn(100000)  # 100k random floats

bit_widths = list(range(2, 32))  # 2 to 32 bits
max_errors = []

for bit in bit_widths:
    q, m, z = quantize(xs, bits=bit)
    max_errors.append(np.max(np.abs(xs - dequantize(q, m, z))))
```

```python
plt.plot(bit_widths, max_errors, marker='o')
plt.xlabel('Bit-width')
plt.ylabel('Max Absolute Error')
plt.title('Quantization Error vs. Bit-width')
plt.show()
```
